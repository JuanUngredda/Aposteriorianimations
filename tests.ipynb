```python
x = np.linspace(0, 10, 500)
y = np.linspace(0, 10, 500)
X, Y = np.meshgrid(x, y)

from scipy.stats import multivariate_normal
rv = multivariate_normal([5, 5], [[1, 0], [0, 1]])
data = np.dstack((X, Y))
z = rv.pdf(data)
obj = plt.contour(x, y, z, np.array([z.min(),np.percentile(z,30), np.percentile(z,70), np.percentile(z,95), z.max()]), color='black')
plt.axis('off')
```

```python
for c in obj.collections[1:]:
    for i in c.get_paths():
        v = i.vertices
        plt.scatter(v[:,0], v[:,1])

plt.show()
```
